```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer

# 1. Create the transformer
trans = PowerTransformer(method='yeo-johnson')

# 2. Fit + transform ApplicantIncome
appIn_trans = trans.fit_transform(data[["ApplicantIncome"]])

# 3. (Optional) Check the shape
print(appIn_trans.shape)   # should be (number_of_rows, 1)

# 4. Convert to Series and check skewness & kurtosis
appIn_trans_series = pd.Series(appIn_trans.reshape(-1,))  # flatten to 1D

print("Skewness:", appIn_trans_series.skew())
print("Kurtosis:", appIn_trans_series.kurtosis())

# 5. Plot histogram of transformed values
sns.histplot(appIn_trans_series)
plt.show()

# 6. Save transformed values into the DataFrame as a new column
data['appIn_trans'] = appIn_trans_series

# 7. Now you can safely cube the transformed column
data['appIn_trans_cube'] = data['appIn_trans'] ** 3
```
